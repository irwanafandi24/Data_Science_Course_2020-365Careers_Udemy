```python
#encode the data using get_dummies
reason_columns = pd.get_dummies(df['Reason for Absence'])
reason_columns.head()
```

```python
reason_columns['check'] = reason_columns.sum(axis=1)
reason_columns.head() #every row, just have 1 value check because based on the data every row only have a single reason
```

```python
print(reason_columns['check'].sum(axis=0)) #same with the number of all data
reason_columns = reason_columns.drop(['check'], axis=1) #delete it, we don't need this data anymore
```

```python
#delete the reason 0, because there isn't reason. Most of employee always attend. Find this fact on Absenteeism Exrcise
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first=True)
reason_columns.head()
```

### Grouping the reason for absence 
- In this section we want add the dummy data into df variable, but wait the feature gonna be too much.
- So we have grouping the data
- Remember if we add the dummy data in df, we must drop Reason for absence column to avoid multicollinearity
<br>
<p>Grouping Data (Internation Classification of Diseases 1-21)</p>
<ol>
    <li>01-14 : Diseases</li>
    <li>15-17 : Pregnancy and giving birth</li>
    <li>18-21 : Poisoning or signs not elsewhere categorized</li>
    <li>22-28 : light reasons for absence</li>
</ol>

```python
df = df.drop(['Reason for Absence'], axis=1)
df.head()
```

```python
reason_columns.columns
```

```python
#get 1-14 column, use max to get the info which rows are categorized in group 1 - 28
rc_1 = reason_columns.iloc[:,:14].max(axis=1)
rc_2 = reason_columns.iloc[:,14:17].max(axis=1)
rc_3 = reason_columns.iloc[:,17:20].max(axis=1)
rc_4 = reason_columns.iloc[:,20:].max(axis=1)
```

```python
#concat the data
df = pd.concat([rc_1, rc_2, rc_3, rc_4, df], axis=1)
```

### Extrac the Day of the week 

```python
# df_reason_mod['Date'][1].day date
df_reason_mod['Date'][0].weekday() #mondoy : 0, 
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
#using df.apply(function) to iterate all data in df and modify by that function
df_reason_mod['Day of the week'] = df_reason_mod['Date'].apply(date_to_weekday)
df_reason_mod.head()
```
