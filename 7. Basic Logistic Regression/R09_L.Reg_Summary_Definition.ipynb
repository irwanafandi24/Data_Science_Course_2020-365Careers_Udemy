```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import seaborn as sns
sns.set()
```

### Load Data

```python
df = pd.read_csv('udemy_data/2.01. Admittance.csv')
df.head()
```

```python
data = df.copy()
data['Admitted']=data['Admitted'].map({'No':0,'Yes':1})
data.head()
```

### Declare Dependent and Independent Var

```python
y = data['Admitted']
x1 = data['SAT']
```

### Logistic Regression

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
result_log = reg_log.fit() #max of Iterations is 35
```

### Summary 

```python
result_log.summary() 
```

### Understand Logistic Regression Summary Tabel
Method : MLE (Maximum Likelihood Estimation)<br>
Likelihood Fuction: is function which estimates how likely it is that the model at handdescribes the real underlaying relationship of the variables. <br>
The bigger likelihood function, the higher probability that our model is correct<br>
MLE tries to maximize the log likelihood<br><br>
Most the value of Log-Likelihood is negative, bigger is better <br><br>
LL-NULL (Log Likelihood Null): the log likelihood of the model which has no independent variables  <br><br>
LLR (Log Likelihood Ratio): same with F-score in linear regression. Measures if our model is statistically different from LL-NULL. We see 5.805e-42 (very low) about 0.0000 so our model is significant. <br><br>
Pseudo R-Square, a good pseudo is somewhere between 0.2 and 0.4. Morover (bahkan) this measure is mostly useful for comparing variation of the same model
